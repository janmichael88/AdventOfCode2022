```python
from collections import defaultdict
data = open('input.txt').readlines()
```

```python
for i in range(len(data)):
    data[i] = data[i].strip('\n')
```

```python
'''
have it return the view as a list
recursive apporach
            * then dp(i,j) = {
                up = 1 + dp(i-1,j) if trees[i][j] > trees[i-1][j]
                down = 1 + dp(i+1,j) if trees[i][j] > trees[i-1][j]
                left = 1 + dp(i,j+1) if trees[i][j] > trees[i][j+1]
                right = 1 + dp(i,j-1) if trees[i][j] > tess[i][j-1]
                return up*down*left*right
'''
```

```python
rows = len(data)
cols = len(data[0])
memo = defaultdict(list)

def dp(i,j):
    #out of bounds, return 1
    #if i were to return 0 the sceneic score would be 0, which cannot be
    #store answers is [up,down,right,left]
    
    if i < 0 or j < 0 or i > rows or j > rows:
        return 1
    #retrieve
    if (i,j) in memo:
        return memo[(i,j)]
    #check up
    if i > 1 and data[i][j] > data[i-1][j]:
        up = 1 + dp(i-1,j)
    else:
        up = 1
    #check down
    if i < rows - 1 and data[i][j] > data[i+1][j]:
        down = 1 + dp(i+1,j)
    else: 
        down = 1
    
    #check right
    if j < cols - 1 and data[i][j] > data[i][j+1]:
        right = 1 + dp(i,j+1)
    else:
        right = 1
        
    #check left
    if j > 1 and data[i][j] > data[i][j-1]:
        left = 1 + dp(i,j-1)
    else:
        left = 1
    
    memo[(i,j)] = sum([up,down,right,left])
    return sum([up,down,right,left])
    
    
        
        
```

```python
dirrs = [(1,0),(-1,0),(0,1),(0,-1)]
for i in range(rows):
    for j in range(cols):
        for dx,dy in dirrs:
            neigh_x = i + dx
            neigh_y = j + dy
            ans = 1
            if (0 <= neigh_x <rows) and (0 <= neigh_y < cols):
                ans *= dp(neigh_x,neigh_y)
            print(ans,dp(i,j))
```
